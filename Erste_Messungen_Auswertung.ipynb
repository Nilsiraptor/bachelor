```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
filenames = [r"..\Data\20211206_130229_lueften.csv",
             r"..\Data\20211206_154423_versuch1.csv",
             r"..\Data\20211206_181218_vergleich1.csv",
             r"..\Data\20211207_145554_versuch2.csv",
             r"..\Data\20211208_092637_versuch3.csv",
             r"..\Data\20211208_122710_versuch4.csv",
             r"..\Data\20211208_171608_versuch5.csv"]
```

```python
versuch1 = np.genfromtxt(filenames[1], skip_header=47, skip_footer=3, delimiter=",")
# co2 = file[:, 3]
# temp = file[:, 5]
x_list1 = np.arange(0, len(versuch1)*60, 60)
```

```python
versuch2 = np.genfromtxt(filenames[3], skip_header=46, skip_footer=3, delimiter=",")
# co2 = file[:, 3]
# temp = file[:, 5]
x_list2 = np.arange(0, len(versuch2)*30, 30)
```

```python
versuch4 = np.genfromtxt(filenames[5], skip_header=46, skip_footer=3, delimiter=",")
# co2 = file[:, 3]
# temp = file[:, 5]
x_list4 = np.arange(0, len(versuch4)*60, 60)
```

```python
versuch5 = np.genfromtxt(filenames[6], skip_header=46, skip_footer=3, delimiter=",")
# co2 = file[:, 3]
# temp = file[:, 5]
x_list5 = np.arange(0, len(versuch5)*5, 5)
```

```python
fig = plt.figure(figsize=(8, 4), dpi=200)
ax = fig.add_subplot(111)

ax.set_xlabel("Zeit in Sekunden")
ax.set_ylabel(r"CO$_2$-Konzentration in ppm")

ax.set_xlim(-50, 2000)
ax.set_ylim(490, 900)
ax.plot(x_list1, versuch1[:, 3], label="Versuch 1")
ax.plot(x_list2, versuch2[:, 3], label="Versuch 2")
ax.plot(x_list4, versuch4[:, 3], label="Versuch 3")
ax.plot(x_list5, versuch5[:, 3], label="Versuch 4")
ax.legend()
fig.tight_layout()
plt.savefig("ErsteVersuche.png", dpi=200)
plt.show()
```
